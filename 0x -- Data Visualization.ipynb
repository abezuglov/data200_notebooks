```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

# Data Visualization Techniques

Essential packages:

* matplotlib
* seaborn

Below we will look at methods for visualizing the data organized by the number of dimensions we will be visualizing and the type of data. We first see how one-dimensional data can be visualized

## One-dimensional data

Let us use the list of Titanic passengers as an example:

```python
df = pd.read_csv('../data/titanic3.csv')
df.head()
```

### Categorical data

Let us visualize how many male and female passengers were on board:

```python
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.countplot(x = 'sex', data = df, ax = ax)
ax.set_title('Number of Titanic passengers by sex')
plt.show()
```

```python
# use countplot to show how many passengers embarked in each port:
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.countplot(x = 'embarked', data = df, ax = ax)
ax.set_title('Number of Titanic passengers by embarked port')
#ax.set_xticks(['S','C','Q'],['Southampton','Cherbourg','Queenston'])
plt.show()
```

```python
# Prepare the data first
df.groupby('embarked').count()['survived']
```

```python
# Draw a pie chart
plt.figure(figsize = (7,7))
ax = plt.subplot(111)
df.groupby('embarked').count()['survived'].plot.pie(
    ax = ax,
    shadow = True,
    explode = [0.1,0.1,0.1],
    x = 'embarked')
ax.set_title('Number of Titanic passengers by embarked port')
plt.show()
```

### One-dimensional continuous data

Here, we can visualize the distributions of fare, and ages of passengers

```python
# use countplot to show how many passengers embarked in each port:
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.distplot(df['fare'].dropna(), ax = ax)
ax.set_title('How much passengers paid for their tickets')
ax.set_xlabel('Fare, pounds')
plt.show()
```

### One-dimensional time series

Let us see the close values of AMZN:

```python
df2 = pd.read_csv('../data/AMZN.csv').set_index('Date')
df2.head()
```

```python
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
df2['Close'].plot(ax = ax)
ax.set_title('AMZN, Close')
ax.set_xlabel('Date')
plt.show()
```

## Two-dimensional data

### Two categorical fields

At this point we can track dependency between two features (fields). Let us start with categorical data and see the dependency between survived and sex of Titanic passengers:

```python
df.head()
```

```python
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.countplot(x = 'sex', hue = 'survived', data = df)
ax.set_title('Number of passengers survived by sex')
plt.show()
```

```python
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.countplot(x = 'sex', hue = 'survived', data = df)
ax.set_title('Number of passengers survived by sex')

leg = ax.get_legend()
leg.set_title("Survived")
legs = leg.texts
legs[0].set_text("No")
legs[1].set_text("Yes")

plt.show()
```

```python
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.countplot(x = 'embarked', hue = 'survived', data = df)
ax.set_title('Number of passengers survived by embarked port')

leg = ax.get_legend()
leg.set_title("Survived")
legs = leg.texts
legs[0].set_text("No")
legs[1].set_text("Yes")

plt.show()
```

### Categorical vs. continuous field

Let us see the dependency between survived and age. This can be drawing two distplots on the same figure:

```python
plt.figure(figsize = (10,5))
ax = plt.subplot(111)
sns.kdeplot(df[df['survived'] == 0]['age'].dropna(), label = 'died')
sns.kdeplot(df[df['survived'] == 1]['age'].dropna(), label = 'survived')
ax.set_title('Survived vs. died based on age')
plt.show()
```

# Data Visualizations with Seaborn

Code from: https://www.kaggle.com/masumrumi/a-statistical-analysis-ml-workflow-of-titanic was used to write the code below

```python
df = pd.read_csv('../data/titanic3.csv')
df.head()
```

```python
pal = {'male':'seagreen', 'female':'lightblue'}

plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "sex", 
            y = "survived", 
            data=df, 
            palette = pal,
            linewidth=2 )
plt.title("Survived/Non-Survived Passenger Gender Distribution")
plt.ylabel("% of passenger survived")
plt.xlabel("Sex")
```

```python
pal = {1:'seagreen', 0:'lightblue'}
sns.set(style='darkgrid')
plt.subplots(figsize = (10,5))
ax = sns.countplot(x = 'sex', 
                   hue='survived',
                   data = df, 
                   linewidth=2, 
                   palette = pal
)

## Fixing title, xlabel and ylabel
plt.title("Passenger Gender Distribution - Survived vs Not-survived")
plt.xlabel("Sex");
plt.ylabel("# of Passenger Survived")

## Fixing xticks
#labels = ['Female', 'Male']
#plt.xticks(sorted(train.Sex.unique()), labels)

## Fixing legends
leg = ax.get_legend()
leg.set_title("Survived")
legs = leg.texts
legs[0].set_text("No")
legs[1].set_text("Yes")
plt.show()
```

```python
plt.subplots(figsize = (10,5))
sns.barplot(x = "pclass", 
            y = "survived", 
            data=df, 
            linewidth=2)
plt.title("Passenger Class Distribution - Survived vs Non-Survived")
plt.xlabel("Socio-Economic class")
plt.ylabel("% of Passenger Survived")
labels = ['Upper', 'Middle', 'Lower']
#val = sorted(train.Pclass.unique())
val = [0,1,2] ## this is just a temporary trick to get the label right. 
plt.xticks(val, labels)
```

```python
fig = plt.figure(figsize=(10,5),)
ax=sns.kdeplot(df.loc[(df['survived'] == 0),'fare'] , color='gray',shade=True,label='not survived')
ax=sns.kdeplot(df.loc[(df['survived'] == 1),'fare'] , color='g',shade=True, label='survived')
plt.title('Fare Distribution Survived vs Non Survived')
plt.ylabel("Frequency of Passenger Survived")
plt.xlabel("Fare")
plt.show()
```

```python
fig = plt.figure(figsize=(10,5),)
ax=sns.kdeplot(df.loc[(df['survived'] == 0),'age'] , color='gray',shade=True,label='not survived')
ax=sns.kdeplot(df.loc[(df['survived'] == 1),'age'] , color='g',shade=True, label='survived')
plt.title('Age Distribution Survived vs Non Survived')
plt.ylabel("Frequency of Passenger Survived")
plt.xlabel("Fare")
plt.show()
```

```python
pal = {1:"seagreen", 0:"gray"}
g = sns.FacetGrid(df,size=5, col="sex", row="survived", margin_titles=True, hue = "survived",
                  palette=pal)
g = g.map(plt.hist, "age", edgecolor = 'white');
g.fig.suptitle("Survived by Sex and Age")
plt.subplots_adjust(top=0.90)
```

```python
g = sns.FacetGrid(df, size=5,hue="survived", col ="sex", margin_titles=True,
                palette=pal,)
g.map(plt.scatter, "fare", "age",edgecolor="w").add_legend()
g.fig.suptitle("Survived by Sex, Fare and Age")
plt.subplots_adjust(top=0.85)
```

```python
sns.factorplot(x = "parch", y = "survived", data = df,kind = "point",size = 8)
plt.title("Factorplot of Parents/Children survived")
plt.subplots_adjust(top=0.85)
```

```python
## heatmeap to see the correlation between features. 
# Generate a mask for the upper triangle (taken from seaborn example gallery)
mask = np.zeros_like(df.corr(), dtype=np.bool)
#mask[np.triu_indices_from(mask)] = True

plt.subplots(figsize = (15,12))
sns.heatmap(df.corr(), 
            annot=True,
            #mask = mask,
            cmap = 'RdBu_r',
            linewidths=0.1, 
            linecolor='white',
            vmax = .9,
            square=True)
plt.title("Correlations Among Features", y = 1.03,fontsize = 20);
```
